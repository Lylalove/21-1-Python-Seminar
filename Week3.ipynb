## list vs Numpy array
list: resizeable, can contain elements of different types.  

numpy array: high-performance multidimentional array object. same type.  
rank : number of dimensions. shape : size of the array along each dimesion. tuple of integers.  

### numpy array 장점
- size : take up less space
- performance : faster than lists
- functionality : have optimized functions such as linear algebra operations built in

ex. dot product, inner prodduct, outer product, ...

for every new element, need another 8 bytes for the reference to the new object. The new integer object itself consumes 28 bytes.  
Size of the list without the size of the elements. 

## 4. Iteration (while, for)

### - while

while (조건) :  
<pre>statements</pre>

```python
i = 0
while i < 5:
    print(i)
    i = i+1
```

```python
condition = False
ind = 0
while not condition:
    print(ind)
    ind = ind + 10
    if (ind > 50) :
        break
```

### - for

set of statements를 실행함. once for each item in a sequence. 

for 인스턴스 in sequence :  
    <pre>statements</pre>

```python
for i in ['a', 1, 'b', 2, 'c', 3]:
    print(i)
```

```python
for i in '1a2b3c':
    print(i)
```

```python
for i in 123:
    print(i)
```

```python
for i in [0, 1, 2, 3, 4, 5]:
    print(i)
```

### - range( )

**range(stop)**
- sequence of numbers starting from 0 to stop-1
- empty sequence if stop is 0 or negative

**range(start, stop [,step])**
- step 이 정의 안되어있다면 default = 1. start ~ stop -1 까지 
- step=0 -> ValueError
- step >= 0 이라면 조건을 만족해야. 만족하지 않으면 empty sequence.  

r[n] = start + step*n (for both positive and negative step)  
where, n >=0 and r[n] < stop (for positive step)  
where, n >= 0 and r[n] > stop (for negative step)

```python
print(list(range(10)))
print(list(range(0)))
print(list(range(-2)))
```

```python
for i in range(6):
    print(i)
```

```python
print(range(6))
print(list(range(6)))
```

```python
print(range(2, 6))
print(list(range(2, 6)))
```

```python
print(range(2, 6, 2))
print(list(range(2, 6, 2)))
# print(range(2, 6, 0))
# print(list(range(2, 6, 0)))
```

```python
print(range(5, 1, -1))
print(list(range(5, 1, -1)))
```

```python
print(range(1, 5, -1))
print(list(range(1, 5, -1)))
```

## 5. 백준 연습문제

### 각각 while 문, for 문을 이용해 구현해보세요!
### [A+B -5] (https://www.acmicpc.net/problem/10952)
### [A+B -4] (https://www.acmicpc.net/problem/10951)
### [더하기 사이클] (https://www.acmicpc.net/problem/1110)

```python
# 1번 while 
cond = False
while not cond:
    inputstr = input()
    inputlist = inputstr.split()
    a = int(inputlist[0])
    b = int(inputlist[1])
    
    if ((a == 0) and (b == 0)):
        cond = True
        break
    print(a+b)
```

```python
# 2번 while 
while True:
    try : 
        inputstr = input()
        if (inputstr == ""):
            break
        inputlist = inputstr.split()
        a = int(inputlist[0])
        b = int(inputlist[1])
        print(a+b)
    except : 
        break
```

```python
N = int(input())
count = 0
newN = N
while True :
    if (newN < 10):
        newN = int(10 * newN  + newN)
    else :
        one = int((newN / 10 + newN % 10) % 10)
        newN = int(10 * (newN % 10) + one)
    count = count + 1
    if newN == N :
        break

print(count)
```
